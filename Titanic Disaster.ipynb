```python
import pandas as pd #data analysis
import numpy as np #numerical operations
import matplotlib.pyplot as plt #data visaulization
plt.rcParams['font.sans-serif']=['SimHei']
```

```python
data_train = pd.read_csv("./train.csv")
data_train
```

```python
data_train.describe()
```

```python
data_train.info()
```

```python
fig = plt.figure(figsize=(9,6))
#fig.set(alpha=0.9)


# Survival Summary
#plt.subplot(3, 5, 1)
plt.subplot2grid((4,5),(0,0)) #sunplots multiple axises in the figure
data_train.Survived.value_counts().plot(kind='bar')# bar chart
plt.title("Survival Summary (1 for survived)") # title
plt.ylabel("Total")  

#Class Distribution
#plt.subplot(3, 5, 3)
plt.subplot2grid((4,5),(0,2))
data_train.Pclass.value_counts(sort=False).plot(kind="bar")
plt.title("Class Distribution")
plt.ylabel("Total")


#Embark Summary
#plt.subplot(3, 5, 15)
plt.subplot2grid((4,5),(0,4))
data_train.Embarked.value_counts().plot(kind='bar')
plt.title("Embark Summary")
plt.ylabel("Total")

#plt.subplot(3, 2, 5)
plt.subplot2grid((5,5),(2,0), colspan=3, rowspan=3)
data_train.Age[data_train.Pclass == 1].plot(kind='kde')   
data_train.Age[data_train.Pclass == 2].plot(kind='kde')
data_train.Age[data_train.Pclass == 3].plot(kind='kde')
plt.xlabel("Age")# plots an axis lable
plt.ylabel("Density") 
plt.title("Class Distribution in Age Group")
plt.legend((u'1st Class', u'2nd Class',u'3rd Class'),loc='best') # sets our legend for our graph.

#Survial Distribution in Age Group
#plt.subplot(3, 5, 5)
plt.subplot2grid((5,5),(2,4), rowspan=3)
plt.scatter(data_train.Survived, data_train.Age, .1)
plt.ylabel("Age")
plt.grid(b=True, which='major', axis='y') 
plt.title("Survial Distribution in Age Group (1 for survived)")
  
plt.show()
```

```python
survived_0 = data_train.Pclass[data_train.Survived == 0].value_counts()
survived_1 = data_train.Pclass[data_train.Survived == 1].value_counts()
print(type(survived_0))
df = pd.DataFrame({'survived':survived_1, 'not_survived':survived_0})
print(df)
df.plot(kind='bar', stacked=True)
plt.title("Survival Summary in Ticket Class")
plt.xlabel("Passenger Ticket Class") 
plt.ylabel("Total") 
plt.show()
```

```python
survived_0 = data_train.Sex[data_train.Survived == 0].value_counts()
survived_1 = data_train.Sex[data_train.Survived == 1].value_counts()
df = pd.DataFrame({'survived':survived_1, 'not_survived':survived_0})
df.plot(kind='bar', stacked=True)
#print(df)
plt.title("Survival Summary in Gender")
plt.xlabel("Gender")
plt.ylabel("Total")

plt.show()
```

```python
fig=plt.figure(figsize=(15,6))
plt.title("Survival Summary in Ticket Class and Gender")
plt.xticks([])
plt.yticks([])

ax1=fig.add_subplot(141)
data_train.Survived[data_train.Sex == 'female'][data_train.Pclass != 3].value_counts(sort=False).plot(kind='bar', label="female highclass", color='red')
ax1.set_xticklabels(["Not Survived", "Survived"], rotation=0)
ax1.legend(["Female/HighClass"], loc='best')

ax2=fig.add_subplot(142, sharey=ax1)
data_train.Survived[data_train.Sex == 'female'][data_train.Pclass == 3].value_counts().plot(kind='bar', label='female, low class', color='pink')
#print(data_train.Survived[data_train.Sex == 'female'][data_train.Pclass == 3].value_counts())
ax2.set_xticklabels(["Not Survived", "Survived"], rotation=0)
plt.legend(["Female/LowClass"], loc='best')

ax3=fig.add_subplot(143, sharey=ax1)
data_train.Survived[data_train.Sex == 'male'][data_train.Pclass != 3].value_counts().plot(kind='bar', label='male, high class',color='steelblue')
#print(data_train.Survived[data_train.Sex == 'male'][data_train.Pclass != 3].value_counts())
ax3.set_xticklabels(["Not Survived", "Survived"], rotation=0)
plt.legend(["Male/HighClass"], loc='best')

ax4=fig.add_subplot(144, sharey=ax1)
data_train.Survived[data_train.Sex == 'male'][data_train.Pclass == 3].value_counts().plot(kind='bar', label='male low class', color='lightblue')
#print(data_train.Survived[data_train.Sex == 'male'][data_train.Pclass == 3].value_counts())
ax4.set_xticklabels(["Not Survived", "Survived"], rotation=0)
plt.legend(["Male/LowClass"], loc='best')

plt.show()
```
