```python
import pandas as pd #data analysis
import numpy as np #numerical operations
import matplotlib.pyplot as plt #data visaulization
plt.rcParams['font.sans-serif']=['SimHei']
```

```python
fig = plt.figure()
x = np.linspace(1, 5, 50)
y1 = 2 * x + 1
plt.plot(x, y1)

fig = plt.figure(figsize=(6,3))
y2 = x ** 2 - 1
plt.plot(x, y2)
plt.plot(x, y1, color="red", linestyle="--")

plt.xlim(0.5, 7)
plt.xlabel("hello world")

new_ticks=np.linspace(0.5, 7, 9)
plt.xticks(new_ticks) # x axis's scales
plt.yticks([25, 12, 8, 2], ["A+", "A", "C", "F"]) # what shows on y axis

plt.show()
```
